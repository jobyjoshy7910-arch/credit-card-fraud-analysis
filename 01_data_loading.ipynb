```python
import pandas as pd
import numpy as np

df=pd.read_csv('creditcard.csv')

print("Shape:",df.shape)
print("\nColumn names:\n",df.columns.tolist())
print("\ndatatypes:\n",df.dtypes)
df.head()
```

```python
print("Missing values:\n",df.isnull().sum())
print("\nFraud distribution:")
print(df['Class'].value_counts())
print(f"\nFraud rate: {df['Class'].mean()*100:.4f}%")
```

```python
print("Transaction amount statistics:")
print(df['Amount'].describe())

print(f"\nMax fraudulent transaction: ${df[df['Class']==1]['Amount'].max()}")
print(f"Max legitimate transaction: ${df[df['Class']==0]['Amount'].max()}")
print(f"\nAvg fraudulent transaction: ${df[df['Class']==1]['Amount'].mean():.2f}")
print(f"Avg legitimate transaction: ${df[df['Class']==0]['Amount'].mean():.2f}")
```
